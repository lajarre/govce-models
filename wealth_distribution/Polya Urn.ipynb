```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
init = pd.Series((np.random.dirichlet(np.ones(10),size=1) * 100).astype(int)[0], dtype='Int64')
rounds = pd.DataFrame([init])
for _ in range(10000):
    last_round = rounds.iloc[len(rounds) - 1].copy()
    color_idx = np.random.choice(len(init), p=last_round/last_round.sum())
    last_round[color_idx] += 1
    rounds = pd.concat([rounds, last_round.to_frame().T], ignore_index=True)
rounds.plot()
plt.title("Polya Urn")
rounds.apply(lambda row: row * 1/row.sum(), axis=1).plot()
plt.title("Polya Urn proportions")
```

```python
init = pd.Series((np.random.dirichlet(np.ones(10),size=1) * 100).astype(int)[0], dtype='Int64')
rounds = pd.DataFrame([init])
for _ in range(10000):
    last_round = rounds.iloc[len(rounds) - 1].copy()
    color_idx = np.random.choice(len(init), p=last_round/last_round.sum())
    last_round[color_idx] += len(rounds)
    rounds = pd.concat([rounds, last_round.to_frame().T], ignore_index=True)
rounds.plot()
plt.title("Polya Urn Increasing")
rounds.apply(lambda row: row * 1/row.sum(), axis=1).plot()
plt.title("Polya Urn Increasing - proportions")
```

```python
init = pd.Series((np.random.dirichlet(np.ones(10),size=1) * 100).astype(int)[0], dtype='Int64')
rounds = pd.DataFrame([init])
for _ in range(100):
    last_round = rounds.iloc[len(rounds) - 1].copy()
    color_idx = np.random.choice(len(init), p=last_round/last_round.sum())
    last_round[color_idx] += int(len(rounds) * last_round[color_idx] / sum(last_round))
    rounds = pd.concat([rounds, last_round.to_frame().T], ignore_index=True)
rounds.plot()
plt.title("Polya Urn Increasing & square proportions")
rounds.apply(lambda row: row * 1/row.sum(), axis=1).plot()
plt.title("Polya Urn Increasing & square proportions - proportions")
```
